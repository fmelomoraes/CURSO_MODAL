ESTRUTURA DE REPETIÇÃO

For - em portugues é para

```python
lista = ['joão', 'jose', 'Andre', 'Marcos']
```

```python
#  Para cada item dentro da lista, imprima o item. Essa automação é conhecida como loop

for item in lista:
 print(item)
```

```python
#  Embora eu tenha usado a palavra item, o Python compreende que estou determinando elementos/itens de uma lista. Assim poderia utilizar qualquer "nome" para descrever esses itens
for proprietario_automovel in lista:
  print(proprietario_automovel)
```

```python
lista_ativos = ['PETRA4.SA', 'WEGW3.SA', 'MGLU3.SA', 'LREN3.SA', 'RENT3.SA', 'VALE3.SA']
lista_ativos
```

```python
for acoes in lista_ativos:
  print(acoes, "é um ativo da B3")
```

```python
#  Automação usando a função break para interromper o loop assim que uma condição for satisfeita (integrando 'for' e 'if')
#  Atenção à indentação correta

for acoes in lista_ativos:
  print(acoes, "é um ativo da B3")
  if acoes == "LREN3.SA":
    break
```

```python
for acoes in lista_ativos:
  print(acoes, "é um ativo da B3")
  if acoes == "WEGW3.SA":
    break
```
